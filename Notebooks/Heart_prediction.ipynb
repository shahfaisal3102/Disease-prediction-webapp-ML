```python
# Importing libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pickle
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading dataset to pandas dataframe

Heart_datasets = pd.read_csv("D:\Disease prediction\Datasets\heart.csv")
```

Spliting Data(Features) from Labels(Target)

```python
X = Heart_datasets.drop(columns='HeartDisease',axis=1)
Y = Heart_datasets['HeartDisease']
```

```python
X
```

```python
Y
```

Label Encoding

```python
# Encoding the categorical data

categorical_cols = X.select_dtypes(include="object").columns.tolist()

label_encoders = {}
for col in categorical_cols:
    label_enc = LabelEncoder()
    X[col] = label_enc.fit_transform(X[col])
    label_encoders[col] = label_enc
```

Spliting data into training and testing set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Model Training

1. Logistic regression

```python
heart_model = LogisticRegression()
```

```python
heart_model.fit(X_train,Y_train)
```

Model Evaluation

1. Accuracy score

```python
# accuracy on training data
X_train_prediction = heart_model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
     

print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on testing data
X_test_prediction = heart_model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
     

print('Accuracy on Testing data : ', testing_data_accuracy)
```

Making Predictive Sysytem

```python
sample_input = pd.DataFrame({
    'Age': [55],
    'Sex': ['M'],
    'ChestPainType': ['ATA'],
    'RestingBP': [140],
    'Cholesterol': [220],
    'FastingBS': [0],
    'RestingECG': ['Normal'],
    'MaxHR': [160],
    'ExerciseAngina': ['N'],
    'Oldpeak': [1.2],
    'ST_Slope': ['Up']
})

# Encode the input using saved label encoders
for col in categorical_cols:
    sample_input[col] = label_encoders[col].transform(sample_input[col])

# make prediction using trained model 
prediction = heart_model.predict(sample_input)
print(prediction)

if(prediction[0] == 0):
    print("Healthy Heart")
else:
    print("Defective Heart")
```
